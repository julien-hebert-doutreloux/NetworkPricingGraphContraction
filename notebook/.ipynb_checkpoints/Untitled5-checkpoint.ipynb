```python
def preamble_sh(cpu, ram, h, m, s, *args):

            # h,m,s must be in forme '00' digit in str
            # args exemple 1
            # "module load python/3.12.0"
            # "source venev/bin/activate"
            # args exemple 2
            # "module load julia"
            # "module load gurobi"
            return [
                    "#!/bin/bash",
                    f"#SBATCH --cpus-per-task={cpu}",
                    f"#SBATCH --mem={ram}G",
                    f"#SBATCH --time={h}:{m}:{s}",
                    "#SBATCH --output=/dev/null",
                    "#SBATCH --partition=optimum"
                    ] + list(args)
args = ["module load julia", "module load gurobi"]
server_time_buffer = 600


command_list=[]
for i, t in enumerate(task,start=1):
    time_limit_sh = 0
    command_list.append([])
    for sub in t:
        
        input_file, output_file = sub
        problem_name = '-'.join(input_file.split('-')[-3:-1:])
        original_file = f'./data/generated/problems/paper/original/000000-000000-{problem_name}-P.json'
        transformation_file = input_file.replace('P.pkl', 'T.pkl')
        
        time_limit = time_dict[problem_name]
        time_limit_sh += (5*(time_limit+10) + 3*30)*sort_fun(sub)
        
        command = f'julia src/julia/script.jl {input_file} {output_file} {time_limit} {original_file} {transformation_file}'
        command_list[-1].append(command)
    time_limit_sh += 1000



    h, m, s = '%02d' % (time_limit_sh // 3600), '%02d' % ((time_limit_sh % 3600) // 60), '00'
    cpu,ram =1, 12


    sh_script = '\n'.join(preamble_sh(cpu, ram, h, m, s, *args) + command_list[-1] +[f'sleep {server_time_buffer}', ])
    with open(f'./special_redo_{i}.sh', 'w') as f:
        f.write(sh_script)
    # print(sh_script)
    #input()
```
