```python
# Méthode de résolution dans l'espace transformé
    # 0. BASELINE - random N in the original NPP (but keep the original M)
    # 1. RANDOM   - random bound for M and N
    # 2. RETRO    - Compute Big M bounds in the transformed problem 
    # 3. MIN      - Use the minimal projection of the bounds given by Big M in the original problem
    # 4. AVG      - Use the average projection of the bounds given by Big M in the original problem
    # 5. MAX      - Use the maximal projection of the bounds given by Big M in the original problem

# Méthode de rétrorésolution
	# 1. Shortest path
		## La solution optimal (t*~) d'un problème transformé est adaptée au problème original et utilisée comme valeur optimale (t*).
	# 2. Lower-bound
		## La solution optimal (t*~) d'un problème transformé est adaptée au problème original et utilisée comme borne inférieure de (t).
	# 3. Upper-bound
		## La solution optimal (t*~) d'un problème transformé est adaptée au problème original et utilisée comme borne supérieure de (t).
	# 4. Comprehensive lower bound
		## Similairement à 2, mais cette la borne inférieur est seulement appliquée lorsqu'elle est inférieure à la borne supérieure N, sinon c'est la borne inférieure par défaut.
        ## Cela fait composante par composante. 
    # 5. Comprehensive upper bound
        ## Similairement à 3, mais cette la borne inférieur est seulement appliquée lorsqu'elle est inférieure à la borne supérieure N, sinon c'est la borne supérieure par défaut.
        ## Cela fait composante par composante. 
	# 6. Hybrid
		## Pour chaque usager ayant un problème transformé réalisable, les conditions de dualité sont remplacées par
		## ("mean(γ(A)' * λ~) <= mean(γ(c) + γ(t))") where mean is the mean function of every element in a given vector
		## ("(c + t)' * x <= b' * γ^-1(λ~)")
		## ("(γ(c) + γ(t))' * x~ <= γ(b)' * λ~")
		## sinon les conditions de dualité originales sont conservées. Aussi, cette méthode fonction seulement lorsque 
        ##  γ est définie par une fonction de projection linéaire.
		## Dans ce travail γ est une projection moyenne.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = result_dict['combined']['balanced_df']
df = df.drop_duplicates()
```

```python
# Analyse de base
# bootstrap n fois les données pour déterminer un moyenne de l'erreur relative.
# Pour toutes méthodes confondues


# Number of bootstrap samples
num_bootstrap_samples = 1500

# List to store bootstrap samples
bootstrap_means = []

# Perform bootstrapping
for _ in range(num_bootstrap_samples):
    bootstrap_sample = df.sample(n=len(df), replace=True)

    # Calculate the mean of 'RE' for each bootstrap sample
    bootstrap_means.append(bootstrap_sample['RE'].mean())

# Convert to a pandas Series for easier analysis
bootstrap_means_series = pd.Series(bootstrap_means)

# Print summary statistics
# print(bootstrap_means_series.describe())

# Plot the distribution of bootstrap means
plt.hist(bootstrap_means, bins=30, edgecolor='k', alpha=0.7)
# plt.title('Distribution of Bootstrap Means for the objective relative error')
plt.xlabel('Mean of objective relative error')
plt.ylabel('Frequency')
plt.show()

# Calculate the 95% confidence interval for the mean
# confidence_interval = np.percentile(bootstrap_means, [2.5, 97.5])
# print(f"95% Confidence Interval for the mean of feature1: {confidence_interval}")


plt.savefig(f'./result/full_combined_bootstrap_{num_bootstrap_samples}.pdf')
plt.show()
plt.close()
```

```python
# bootstrap 1000 fois les données pour déterminer un moyenne de l'erreur relative.
# Pour toute chaque méthode séparément

# Number of bootstrap samples
num_bootstrap_samples = 1500

# Separate the dataset by class
classes = df['Class1'].unique()
bootstrap_means = {cls: [] for cls in classes if '-x' not in cls}

# Perform bootstrapping for each class
for cls in bootstrap_means:
    class_data = df[df['Class1'] == cls]
    for _ in range(num_bootstrap_samples):
        bootstrap_sample = class_data.sample(n=len(class_data), replace=True)
        bootstrap_means[cls].append(bootstrap_sample['RE'].mean())



# Convert to pandas Series for easier analysis
bootstrap_means_series = {cls: pd.Series(means) for cls, means in bootstrap_means.items()}

# Print summary statistics for each class
# for cls in classes:
#     print(f"Summary statistics for class {cls}:")
#     print(bootstrap_means_series[cls].describe())
#     print()

# Plot the distribution of bootstrap means for each class
rows = {}
columns={}
for cls in bootstrap_means:
    _, option, heuristic = cls.split('-')
    
    if not option in rows:
        rows[option] = len(rows) + 1
        rows[option] -= 1
        
    if not heuristic in columns:
        columns[heuristic] = len(columns) + 1
        columns[heuristic] -= 1
        

coordinates = {cls: (rows[cls.split('-')[1]], columns[cls.split('-')[2]]) for cls in bootstrap_means}



# Create a figure and a set of subplots
fig, axs = plt.subplots(len(rows), len(columns), figsize=(25, 50), sharey=True, sharex=True)


for cls in bootstrap_means:
    H4, option, heuristic = cls.split('-')
    i, j = coordinates[cls]
    
    axs[i, j].hist(bootstrap_means_series[cls], bins=30, edgecolor='k', alpha=0.7, label=f'H4={H4}')
    axs[i, j].set_title(f'Class {option}-{heuristic}')
    axs[i, j].legend()

    if j==0:
        axs[i, j].set_ylabel('Frequency')
    
    if i==len(rows)-1:
        axs[i, j].set_xlabel('Mean of objective relative error')

plt.tight_layout()


# Calculate the 95% confidence interval for the mean for each class
# confidence_intervals = {cls: np.percentile(means, [2.5, 97.5]) for cls, means in bootstrap_means.items()}
# for cls in bootstrap_means:
#     print(f"95% Confidence Interval for the mean of feature1 (Class {cls}): {confidence_intervals[cls]}")

plt.savefig(f'./result/full_seperated_bootstrap_{num_bootstrap_samples}.pdf')
plt.show()
plt.close()
```

```python
# Boxplot du ratio d'optimalité vs les méthodes de rétrorésolution
# pour toutes les méthodes de résolution dans l'espace transformé.



##########################################  ##########################################
# RND - Shortest path          (RND - 1) #  # ZIP - Shortest path          (ZIP - 1) #
# RND - Lower bound            (RND - 2) #  # ZIP - Lower bound            (ZIP - 2) #
# RND - Upper bound            (RND - 3) #  # ZIP - Upper bound            (ZIP - 3) #
# RND - Adaptative lower bound (RND - 4) #  # ZIP - Adaptative lower bound (ZIP - 4) #
# RND - Adaptative upper bound (RND - 5) #  # ZIP - Adaptative upper bound (ZIP - 5) #
# RND - Comparative projection (RND - 6) #  # ZIP - Comparative projection (ZIP - 6) #
##########################################  ##########################################
 
##########################################  ##########################################
# MIN - Shortest path          (MIN - 1) #  # AVG - Shortest path          (AVG - 1) #
# MIN - Lower bound            (MIN - 2) #  # AVG - Lower bound            (AVG - 2) #
# MIN - Upper bound            (MIN - 3) #  # AVG - Upper bound            (AVG - 3) #
# MIN - Adaptative lower bound (MIN - 4) #  # AVG - Adaptative lower bound (AVG - 4) #
# MIN - Adaptative upper bound (MIN - 5) #  # AVG - Adaptative upper bound (AVG - 5) #
# MIN - Comparative projection (MIN - 6) #  # AVG - Comparative projection (AVG - 6) #
##########################################  ##########################################

                        ##########################################
                        # MAX - Shortest path          (MAX - 1) #
                        # MAX - Lower bound            (MAX - 2) #
                        # MAX - Upper bound            (MAX - 3) #
                        # MAX - Adaptative lower bound (MAX - 4) #
                        # MAX - Adaptative upper bound (MAX - 5) #
                        # MAX - Comparative projection (MAX - 6) #
                        ##########################################

hue_order = ['rnd','bsl','zip','min','avg','max']
order = ['1','2','3','4','5','6','x']
xticklabels = [
                'Shortest path (1)', 
               'Lower bound (2)', 
               'Upper bound (3)', 
               'Adapt. lower bound (4)', 
               'Adapt. upper bound (5)', 
               'Hybrid (6)', 
               'Contracted space'
]


for yy in ['adjusted_time_ratio', 'optimal_ratio']:
    for i in [0, 1]:
        fig, ax = plt.subplots(figsize=(15, 10))
        sns.boxplot(data=df[df['H4']==i], x="heuristic", y=yy, hue="option", hue_order=hue_order, order=order, ax=ax).set(xlabel='')
        
        ax.tick_params(which='major', width=1.0, length=30, axis='x')
        ax.set_xticklabels(xticklabels)
        plt.setp(ax.get_xticklabels(), rotation=45)
        
        ax.axhline(y=1, color='red', linestyle='--')
        ax.axhline(y=0.5, color='red', linestyle='--')
        ax.axhline(y=0.75, color='red', linestyle='--')
        
        ax.set_ylabel(yy.replace("_", " ").title(), fontsize=16)
        
        # ax.legend(bbox_to_anchor=(1.025, 1), loc='upper left', fontsize=14)
        ax.legend(bbox_to_anchor=(0.5, 1.025), loc='lower center', ncol=len(hue_order), fontsize=14)
        
        plt.tick_params(axis='both', which='major', labelsize=14)
        
        
        # To set the x-axis range
        if yy == 'adjusted_time_ratio':
            plt.ylim(ymin=-0.05, ymax=2.5)
        elif yy=='optimal_ratio':
            plt.ylim(ymin=-0.05, ymax=1.3)
            
        plt.tight_layout()
        plt.savefig(f'./result/boxplot_full_{yy}_H4_{i}.pdf')
        plt.show()
        plt.close()
```
